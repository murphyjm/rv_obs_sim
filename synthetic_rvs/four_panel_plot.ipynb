```python
import os
import numpy as np
import pandas as pd
from utils import plot_four_panel_recovery_test_results
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2
```

```python
def generate_all_plots(savefig=False):
    parent_dir = 'completed_runs_nsim_3000'

    for tel in ['hires', 'kpf']:
        for astro in ['quiet', 'active']:
            fig, axes, contours, scatters = plot_four_panel_recovery_test_results(parent_dir, 
                                                                                  tel, 
                                                                                  astro, 
                                                                                  savefig=savefig, 
                                                                                  red_noise=False)
            # Also create the plots for the red noise case
            if tel == 'kpf' and astro == 'quiet':
                fig, axes, contours, scatters = plot_four_panel_recovery_test_results(parent_dir, 
                                                                                      tel, 
                                                                                      astro, 
                                                                                      savefig=savefig, 
                                                                                      red_noise=True)
```

```python
generate_all_plots(savefig=True)
```
